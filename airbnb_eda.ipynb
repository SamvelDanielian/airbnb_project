## Data Cleaning

In this section, we will perform data cleaning to handle missing values, remove invalid observations.

```python
import pandas as pd
import numpy as np

# 1) Checking missing values BEFORE
print("Missing values BEFORE:")
display(df.isna().sum().sort_values(ascending=False).head(20))

# 2) Filling numeric columns with median
numeric_cols_to_fill = [
    'Guest Satisfaction',
    'Cleanliness Rating',
    'Bedrooms',
    'Person Capacity',
    'Price'
]

for col in numeric_cols_to_fill:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# Filling categorical columns with mode
categorical_cols_to_fill = [
    'Day',
    'Room Type',
    'Superhost',
    'Multiple Rooms',
    'Private Room',
    'Shared Room'
]

for col in categorical_cols_to_fill:
    if col in df.columns:
        # Taking the first most frequent
        df[col] = df[col].fillna(df[col].mode(dropna=True)[0])

# Checking missing values AFTER
print("\nMissing values AFTER:")
display(df.isna().sum().sort_values(ascending=False).head(20))

# Doing basic sanity checks
print("\nShape after filling missing values:", df.shape)
display(df.head())
```

### Missing Values Imputation

Missing values were handled separately for numerical and categorical features. For numerical variables, missing values were filled using the median. The median was chosen instead of the mean because it is more robust to extreme values and better represents the central tendency in skewed distributions. For categorical variables, missing values were filled using the mode (most frequent value), which preserves the most common category in each feature.

After this step, the dataset contains no missing values and is suitable for further cleaning and feature engineering.

```python
df = df[df['Price'] > 0]
df = df[df['Person Capacity'] > 0]
```

### Removing Invalid Observations

Listings with non-positive prices or person capacity were removed, as such values are not meaningful in a real-world Airbnb context.

```python
price_q99 = df['Price'].quantile(0.99)
df = df[df['Price'] <= price_q99]
```

### Handling Price Outliers

To reduce the impact of extreme price values, listings with prices above the 99th percentile
were removed. This approach preserves most of the data while limiting the influence of outliers.

```python
df.isnull().sum()
df.shape
```

## Feature Engineering

In this section, new features are created to enhance the dataset and extract more meaningful insights from the existing variables.

### Adding Price Per Person Feature

The strongest feature is here! Here we divide the price by the number of guests to get a fair comparison between apartments of different sizes.

```python
df['Price per Person'] = df['Price'] / df['Person Capacity']
df.shape #Here we got a new column and made 20!
```

### Adding Above City Average Price Feature

Another astonishing feature, that is missing, is Above City Average Price. We will check whether a given Airbnb is more expensive than the average for its city or not.

```python
city_avg_price = df.groupby('City')['Price'].mean()

df['Above City Average'] = df.apply(
    lambda row: row['Price'] > city_avg_price[row['City']],
    axis=1
)

df
```

### Adding City Distance Category Feature

Here we have a new feature which will describe if the current Airbnb has following values from city center: central, medium distance, far.

```python
def distance_category(dist):
    if dist <= 2:
        return 'Central'
    elif dist <= 5:
        return 'Medium Distance'
    else:
        return 'Far from Center'

df['Distance Category'] = df['City Center (km)'].apply(distance_category)

df.head() 
```

### Adding Value for Money Feature

Here we calculate if the Airbnb deserves its cost.

```python
avg_price_per_person = df['Price per Person'].mean()

df['Value for Money'] = df['Price per Person'].apply(
    lambda x: 'Good Value' if x <= avg_price_per_person else 'Expensive'
)
```

## Cleaning Part

### Removing Unnecessary Business Column

The business column does not add analytical value to the problem at hand. Therefore, we are removing it to make the dataset cleaner and more purposeful.

```python
col_to_drop = "Business"

print("Before drop: Business in columns?", col_to_drop in df.columns)

# Dropping
df = df.drop(columns=[col_to_drop], errors="ignore")

print("After drop: Business in columns?", col_to_drop in df.columns)
print("Current shape:", df.shape)

# Showing all final columns
display(df.columns)
```
